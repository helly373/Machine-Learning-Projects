```python
df = df[['coaster_name',
    # 'Length', 'Speed',
    'Location', 'Status',
    # 'Opening date', 'Type',
    'Manufacturer',
    # 'Height restriction', 'Model', 'Height',
    #    'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
    #    'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
    #    'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
    #    'Track layout', 'Fastrack available', 'Soft opening date.1',
    #    'Closing date',
    # 'Opened',
    # 'Replaced by', 'Website',
    #    'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
    #    'Single rider line available', 'Restraint Style',
    #    'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
    'latitude', 'longitude',
    'Type_Main',
       'opening_date_clean',
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
    # 'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.dtypes
```

```python
df['opening_date_clean']  = pd.to_datetime(df['opening_date_clean'])
```

```python
#Renaming the columns
df = df.rename(columns={'coaster_name': 'Coaster_Name',
                   'year_introduced': 'Year_Introduced',
                   'opening_date_clean': 'Opening_Date',
                   'speed_mph': 'Speed_mph',
                   'height_ft': 'Height_ft',
                   'Inversions_clean': 'Inversions',
                   'Gforce_clean': 'Gforce',
                   })
```

```python
df.columns
```

```python
#'~' is for inversing. So in this sentences, we removed the duplicate rows by 3 features by inversing and resetting the index
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True).copy()
```

```python
df.shape
```

```python
ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title = 'Top 10 Years Coasters Introduced')
ax.set_xlabel('Year Introduced')
ax.set_ylabel('Number of Coasters')
```

```python
ax = df['Speed_mph'].plot(kind='hist', bins=50, title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed mph')
```

```python
ax = df['Speed_mph'].plot(kind='kde', title = 'Coaster Speed (mph)')
ax.set_xlabel('Speed mph')
```

```python
df.plot(kind='scatter', x='Speed_mph', y='Height_ft', title="Coasters' Speed VS Height")
plt.show()
```

```python
#compared 2 features with each other
ax = sns.scatterplot(x='Speed_mph', y='Height_ft', data=df, hue='Year_Introduced')
ax.set_title("Coaster Speed (mph) VS Height (ft)")
```

```python
#coparing multiple features with each other
sns.pairplot(df, vars=['Year_Introduced', 'Height_ft', 'Speed_mph', 'Gforce', 'Inversions'], hue='Type_Main')
plt.show()
```

```python
#looking at the correlation
df_corr = df[['Year_Introduced', 'Height_ft', 'Speed_mph', 'Gforce', 'Inversions']].dropna().corr()
print(df_corr)
```

```python
sns.heatmap(df_corr, annot=True)
plt.show()
```

```python
df.query('Location != "Other"').groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title="Average Coaster Speed By Location")
plt.show()
```
